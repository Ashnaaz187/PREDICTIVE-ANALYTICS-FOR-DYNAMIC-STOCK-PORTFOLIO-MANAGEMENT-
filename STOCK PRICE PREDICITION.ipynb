```python
#Training and Testing 
data_training=pd.DataFrame(df['Close'][0:int(len(df)*0.70)])
data_testing=pd.DataFrame(df['Close'][int(len(df)*0.70): int(len(df))])
```

```python
data_training.shape
```

```python
data_testing.shape
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler(feature_range=(0,1))
```

```python
data_training_array=scaler.fit_transform(data_training)
```

```python
data_training_array
```

```python
x_train=[]
y_train=[]

for i in range(100, data_training_array.shape[0]):
    x_train.append(data_training_array[i-100:i])
    y_train.append(data_training_array[i, 0])

x_train, y_train =np.array(x_train), np.array(y_train)
```

```python
x_train.shape
```

```python
pip install keras
```

```python
!pip install tensorflow
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input,Dense, Dropout, LSTM
```

"LSTM Input -> 3D Array(batch_size,time_steps,seq_len)LSTM 2D OR 3D 2D->(batch_sizem units) 3D->(batch_size,time_steps,units)"

```python
model = Sequential([
    Input(shape=(x_train.shape[1], 1)),             
    LSTM(units=50, activation='relu', return_sequences=True),
    Dropout(0.2),
    LSTM(units=60, activation='relu', return_sequences=True),
    Dropout(0.3),
    LSTM(units=80, activation='relu', return_sequences=True),
    Dropout(0.4),
    LSTM(units=120, activation='relu'),
    Dropout(0.5),
    Dense(units=1)
])
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='mean_squared_error')
model.fit(x_train,y_train,epochs=150)
```

```python
past_100_days=data_training.tail(100)
```

```python
import pandas as pd

final_df = pd.concat([past_100_days, data_testing], ignore_index=True)
```

```python
final_df.head()
```

```python
input_data=scaler.fit_transform(final_df)
```

```python
x_test=[]
y_test=[]

for i in range(100, input_data.shape[0]):
    x_test.append(input_data[i-100:i])
    y_test.append(input_data[i, 0])

x_test, y_test =np.array(x_test), np.array(y_test)
```

```python
x_test.shape
```

```python
y_predicted=model.predict(x_test)
```

```python
y_predicted.shape
```

```python
scaler.scale_
```

```python
scaler_factor = 1 / 0.00133053
y_predicted = y_predicted * scaler_factor   
y_test = y_test * scaler_factor             
```

```python
plt.figure(figsize=(8,4))
plt.plot(y_test, label='Original Price')
plt.plot(y_predicted, label='Predicted Price')
plt.legend()
plt.grid(True)
plt.show()
```

```python
scaler_factor = 1 / 0.00729727
y_predicted = y_predicted * scaler_factor   
y_test = y_test * scaler_factor   
```

```python
from sklearn.metrics import accuracy_score

# Actual labels
y_true = [0, 1, 1, 0, 1, 0, 1]

# Predicted labels by your code/model
y_pred = [0, 0, 1, 0, 1, 0, 1]

# Calculate accuracy
accuracy = accuracy_score(y_true, y_pred)
print(f"Accuracy: {accuracy*100:.2f}%")
```

```python
import numpy as np
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report

# Convert continuous values into binary classes
# 1  price increased; 0  price decreased or same
y_test_class = np.where(np.diff(y_test.flatten(), prepend=y_test[0]) > 0, 1, 0)
y_pred_class = np.where(np.diff(y_predicted.flatten(), prepend=y_predicted[0]) > 0, 1, 0)

# Compute confusion matrix
cm = confusion_matrix(y_test_class, y_pred_class)

# Display confusion matrix
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(cmap='Blues')
```

```python
print(classification_report(y_test_class, y_pred_class))
```
