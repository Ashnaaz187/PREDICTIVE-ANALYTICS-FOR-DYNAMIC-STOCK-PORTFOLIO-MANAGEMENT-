```python
import yfinance as yf
import plotly.graph_objects as go

# Download stock data
stock = 'AAPL'
data01 = yf.download(stock, start='2020-01-01', end='2023-01-01')

# Create candlestick chart
fig = go.Figure(data=[
    go.Candlestick(
        x=data01.index,  # Use index as Date
        open=data01['Open'],
        high=data01['High'],
        low=data01['Low'],
        close=data01['Close']
    )
])

# Customize layout
fig.update_layout(
    title=f'{stock} Candlestick Chart',
    xaxis_title='Date',
    yaxis_title='Price (USD)',
    xaxis_rangeslider_visible=False
)

# Show the figure
fig.show()

```

```python
import yfinance as yf

stock = 'AAPL'
data01 = yf.download(stock, start='2020-01-01', end='2023-01-01')

plt.figure(figsize=(12,6))
plt.plot(data01['Open'], label=f'{stock} Opening Price', linewidth=1)
plt.title(f'{stock} Opening price over time')
plt.xlabel('Date')
plt.ylabel('Opening Price (USD)')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.plot(data01['Close'], label=f'{stock} Closing Price', color='green', linewidth=1)
plt.title(f'{stock} Closing Price Over Time')
plt.xlabel('Date')
plt.ylabel('Closing Price (USD)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.plot(data01['High'], label=f'{stock} High Price', color='orange', linewidth=1)
plt.title(f'{stock} High Price Over Time')
plt.xlabel('Date')
plt.ylabel('High Price (USD)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.plot(data01['Low'], label=f'{stock} Low Price', color='red', linewidth=1)
plt.title(f'{stock} Low Price Over Time')
plt.xlabel('Date')
plt.ylabel('Low Price (USD)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.plot(data01['Volume'], label=f'{stock} Volume', color='purple', linewidth=1)
plt.title(f'{stock} Trading Volume Over Time')
plt.xlabel('Date')
plt.ylabel('Volume')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

stock = 'AAPL'
data = yf.download(stock, start='2020-01-01', end='2027-01-01')

ma100 = data['Close'].rolling(100).mean()

plt.figure(figsize=(12,6))
plt.plot(data['Close'], label='Closing Price')
plt.plot(ma100, label='100-Day Moving Average', color='orange')
plt.title(f'{stock} Closing Price and 100-Day MA')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

stock = 'AAPL'
data = yf.download(stock, start='2020-01-01', end='2027-01-01')

ma200 = data['Close'].rolling(200).mean()

plt.figure(figsize=(12,6))
plt.plot(data['Close'], label='Closing Price')
plt.plot(ma200, label='200-Day Moving Average', color='orange')
plt.title(f'{stock} Closing Price and 200-Day MA')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.plot(ma100, label=f'{stock} Volume', color='purple', linewidth=2)
plt.plot(ma200, label=f'{stock} Volume', color='yellow', linewidth=2)
plt.title(f'{stock} 100ma vs 200ma')

plt.legend()
plt.grid(True)
plt.show()
```
